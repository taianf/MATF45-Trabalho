```r
#carregando o pacote tm
require(tm)
require(slam)
require(wordcloud2)
require(RColorBrewer)
```

```r
#Lendo os arquivos do diretorio para criar o Corpus
arquivos <- DirSource(directory = "Livros/",encoding ="UTF-8")
arquivos
```

```r
#ler os arquivos em pdf
corpus <- VCorpus(arquivos, 
                  readerControl = list(reader = readPDF))
```

```r
#verificando as informacoes 
corpus
```

```r
#visualizando o texto do primeiro arquivo
corpus[[1]]$content[1:2]
```

```r
#Funcao para remover um marcador
removeMarcador <- function(texto,marcador="<[^>]*>"){
  #por padrao, remove codigo html "<[^>]*>"
  texto <- gsub(pattern=marcador,replacement=" ",x=texto)
  
  return(PlainTextDocument(texto))
  
}
```

```r
#Alterar tudo para minusculo
corpus[[1]]$content[2]
corpus <- tm_map(corpus, content_transformer(tolower))
corpus[[1]]$content[2]
```

```r
#Removendo espaços em brancos que podem aparecer
corpus <- tm_map(corpus,stripWhitespace)
corpus[[1]]$content[2]
```

```r
#Removendo stopwords
corpus[[1]]$content[2]
corpus <- tm_map(corpus,removeWords,c(stopwords("portuguese"),"the","and"))
corpus[[1]]$content[2]
```

```r
#Removendo um caracter em especial
corpus <- tm_map(corpus,content_transformer(removeMarcador),"“")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"”")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"”")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"—")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"‘")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"’")
corpus <- tm_map(corpus,content_transformer(removeMarcador),"[0-9]+")
```

```r
#Removendo pontuacao
corpus<- tm_map(corpus,removePunctuation)
corpus[[1]]$content[2]
```

```r
#Matriz de frequencia
matriz_termos <- TermDocumentMatrix(corpus)
```

```r
#inspecionando a matriz
inspect(matriz_termos[1:7,1:4])
```

```r
#Temos com frequencia de 5
lista <- findFreqTerms(matriz_termos, lowfreq = 100)
lista
```

```r
#Associacao
findAssocs(matriz_termos, "sucesso", 0.9)
```

```r
#otimiza o processamento
matriz1 = rollup(matriz_termos[,1], 2, na.rm=TRUE, FUN = sum)
matriz2 = rollup(matriz_termos[,2], 2, na.rm=TRUE, FUN = sum)
matriz3 = rollup(matriz_termos[,3], 2, na.rm=TRUE, FUN = sum)
matriz4 = rollup(matriz_termos[,4], 2, na.rm=TRUE, FUN = sum)
```

```r
matriz1 <- as.matrix(matriz1)
matriz2 <- as.matrix(matriz2)
matriz3 <- as.matrix(matriz3)
matriz4 <- as.matrix(matriz4)
matriz1[1:10,]
matriz2[1:10,]
matriz3[1:10,]
matriz4[1:10,]

```

```r
#ordenando a lista a partir da frequencia dos termos
palavras1 <- sort(rowSums(matriz1),decreasing=TRUE)
palavras2 <- sort(rowSums(matriz2),decreasing=TRUE)
palavras3 <- sort(rowSums(matriz3),decreasing=TRUE)
palavras4 <- sort(rowSums(matriz4),decreasing=TRUE)
```

```r
#exibindo as 10 primeiras linhas
palavras1[1:10]
palavras2[1:10]
palavras3[1:10]
palavras4[1:10]
```

```r
#criando o data.frame com as colunas word e freq
palavras1 <- data.frame(word = names(palavras1),freq=palavras1)
palavras2 <- data.frame(word = names(palavras2),freq=palavras2)
palavras3 <- data.frame(word = names(palavras3),freq=palavras3)
palavras4 <- data.frame(word = names(palavras4),freq=palavras4)
```
